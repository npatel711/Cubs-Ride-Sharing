```python
#Import dependencies

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sqlalchemy import create_engine
```

```python
#Read data files

hawks = pd.read_csv("Resources/Blackhawks 2018.csv")
bulls = pd.read_csv("Resources/Bulls 2018.csv")

#hawks.head()
bulls.head()
```

```python
new_hawks = hawks.rename (columns={'Time': 'Eastern', 'Time-1': 'Time', "Unnamed: 4": "Vs", "Unnamed: 8": "W/L", 'Unnamed: 9': 'OT'})
new_hawks.head()
```

```python
drop_hawks = ['GP', 'Eastern', 'GF', 'GA', 'W', 'L', 'OL', 'Streak']
new_hawks.drop(columns = drop_hawks, inplace = True)
new_hawks.head()
```

```python
new_bulls = bulls.rename (columns={'Start (CT)': 'Time', "Unnamed: 6": "Vs", "Unnamed: 8": "W/L"})
new_bulls.head()
```

```python
drop_bulls = ['G', 'Date- Old', 'Start (ET)', 'Unnamed: 5', 'Unnamed: 9', 'Tm',
          'Opp', 'W', 'L', 'Streak', 'Notes']
new_bulls.drop(columns = drop_bulls, inplace = True)
new_bulls.head()
```

```python
#Join data files

new_hawks.append(new_bulls)
```

```python
#Merge new data
UC_games = new_hawks[["Date","Time","Vs","Opponent", "W/L", "LOG", "Att."]]
UC_games.head()

#Save to a new csv file
UC_games.to_csv("Output/UC Games.csv")
```

```python
database_path = "etl"
engine = create_engine(f"sqlite:///{database_path}")
```

```python
engine.table_names()
```

```python
UC_games.to_sql(name='UCGames', con=engine, if_exists='append', index=False)
```
